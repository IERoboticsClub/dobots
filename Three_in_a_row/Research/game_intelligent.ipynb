```python
import random
import copy

# tic tac toe board with all cells initially set to 0
result = [[2, 0, 2],
          [1, 1, 2], 
          [0, 0, 1]]
print(result)

# count of how many tokens you have placed on the board
tokens_placed = 4

count = 0

def get_square_number(row, col):
    # row and col are 0-indexed, so add 1 to convert to 1-indexed square number
    return row * 3 + col + 1

def get_square_new_number(row, col):
    # row and col are 0-indexed, so add 1 to convert to 1-indexed square number
    return row * 3 + col + 1


# Checking if there is a winning move avaiable 
def check_winning_move(player, result):
    # check rows
    for row in result:
        if row.count(player) == 2 and row.count(0) == 1:
            col = row.index(0)
            return (result.index(row), col)
        
    # check columns
    for i in range(3):
        col = [result[j][i] for j in range(3)]
        if col.count(player) == 2 and col.count(0) == 1:
            row = col.index(0)
            return (row, i)
        
    # check diagonals
    diagonal1 = [result[i][i] for i in range(3)]
    diagonal2 = [result[i][2-i] for i in range(3)]
    if diagonal1.count(player) == 2 and diagonal1.count(0) == 1:
        index = diagonal1.index(0)
        return (index, index)
    if diagonal2.count(player) == 2 and diagonal2.count(0) == 1:
        index = diagonal2.index(0)
        return (index, 2-index)
    
    return None

# check if there is a final winning move from the opponent and blocks it
def check_blocking_move(player, result):
    return check_winning_move(3 - player, result)


def game_over(board):
    # Check rows
    for row in board:
        if len(set(row)) == 1 and row[0] != 0:
            return True

    # Check columns
    for col in range(3):
        if len(set([board[row][col] for row in range(3)])) == 1 and board[0][col] != 0:
            return True

    # Check diagonals
    if len(set([board[i][i] for i in range(3)])) == 1 and board[0][0] != 0:
        return True
    
    if len(set([board[i][2-i] for i in range(3)])) == 1 and board[0][2] != 0:
        return True

    return False

def winning_sequence(result, player):
    # Check if this token is part of a two-out-of-three sequence token to get a winning move
    result_copy = copy.deepcopy(result)
    for i in range(3):
        for j in range(3):
            if result_copy[i][j] == 0:
                result_copy[i][j] = player  # place token of current player on this position
                if game_over(result_copy) == True:
                    return True
                result_copy[i][j] = 0  # reset this position to empty
    
    return False

def blocking_position(result, row, col, player):
    # Check if this token is blocking the enemy from doing a winning move
    result_copy = copy.deepcopy(result)
    result_copy[row][col] = 3 - player  # place token of enemy player on this position
    if game_over(result_copy) == True:
        return True
    
    return False


def is_blocking_or_winning_position(result, row, col, player):
    # Check if this token is blocking the enemy from doing a winning move
    result_copy = copy.deepcopy(result)
    result_copy[row][col] = 3 - player  # place token of enemy player on this position
    if game_over(result_copy) == True:
        return True
    
    # Check if this token is part of a two-out-of-three sequence token to get a winning move
    result_copy = copy.deepcopy(result)
    for i in range(3):
        for j in range(3):
            if result_copy[i][j] == 0:
                result_copy[i][j] = player  # place token of current player on this position
                if game_over(result_copy) == True:
                    return True
                result_copy[i][j] = 0  # reset this position to empty
    
    return False

# place three tokens on the board at random locations
# place token on the board at random location

# MAIN
if tokens_placed < 3:
    # choose a random row and column
    # check for a winning move for the player with tokens value 1
    winning_move = check_winning_move(1, result)
    if winning_move is not None:
        row, col = winning_move
        result[row][col] = 2
        print(get_square_number(row, col))

    else:
        # check for a blocking move for the player with tokens value 1
        blocking_move = check_blocking_move(1, result)
        if blocking_move is not None:
            row, col = blocking_move
            result[row][col] = 2
            print(get_square_number(row, col))

        else:
            while True:
                row = random.randint(0, 2)
                col = random.randint(0, 2)
                # check if the cell is empty
                if result[row][col] == 0:
                    break

            # place your token in the cell
            result[row][col] = 2

            position = get_square_number(row, col)
            print(position)

    tokens_placed += 1


# if all tokens have been placed, move one to a different location
else:

    # Loops through the cells and if there is a token that is not blocking an enemies winning move or its not an almost winning-move for itself, 

    validation = False
    validation2 = False
    validation3 = False

    # If there is a winning sequence take the token that is not in the sequence.
    n_of_winning_sequences = 0

    # If there is a winning sequence take the token that is not in the sequence.
    for row in range(3):
        for col in range(3):
            if result[row][col] == 2:
                if winning_sequence(result,2) == True:
                    n_of_winning_sequences += 1
                if winning_sequence(result, 2) == False: 
                    if n_of_winning_sequences != 0:
                        validation = True
                        break

        if validation == True:
            break
    
    # Check if the code has run to the last cell of the board
    if row == 2 and col == 2:
        if result[row][col] == 2:

            # If the last token is not in a winning sequence but the other two tokens are, take the last token.
            if winning_sequence(result, 2) == False: 
                if n_of_winning_sequences != 0:
                    pass

            # If the last token is in a winning sequence but the other two tokens are too, take the last token.
            if winning_sequence(result, 2) == True:
                if n_of_winning_sequences != 0:
                    pass
            if winning_sequence(result, 2) == False:
                if n_of_winning_sequences == 0:
                    for row in range(3):
                        for col in range(3):
                            if result[row][col] == 2:
                                if blocking_position(result, row, col, 2) == False:
                                    validation2 = True
                                    break

                        if validation2 == False:
                            break
                    
                    if row == 2 and col == 2:
                        while True:
                            row = random.randint(0, 2)
                            col = random.randint(0, 2)
                            if result[row][col] == 2:
                                break
                    


        # If the last cell is not a 2, it means there are no winning sequences posible so we look for not taking tokens in blooking positions.
        if result[row][col] != 2:
            for row in range(3):
                for col in range(3):
                    if result[row][col] == 2:
                        if blocking_position(result, row, col, 2) == False:
                            validation3 = True
                            break

                if validation3 == False:
                    break
            
            if row == 2 and col == 2:
                while True:
                    row = random.randint(0, 2)
                    col = random.randint(0, 2)
                    if result[row][col] == 2:
                        break




    

    # If all the tokens are blocking a move or are almost formating a winning move choose the one that is blocking so you can put it 
    # in the winning cell. If they are no blocking moves just choose a random cell

    if row == 2 and col == 2:
        
        # If the last token is 2 check if its in a is_blocking_or_winning_position, if it is check for tokens that are in a blocking position
        # because it's better to pick up a token that is in a clocking position that one in a winning position.
        # If there are none take a random token.

        if result[row][col] == 2:
            if is_blocking_or_winning_position(result, row, col, 2) == True:
                
                    for row in range(3):
                        for col in range(3):
                            if result[row][col] == 2:
                                result_copy = copy.deepcopy(result)
                                result_copy[row][col] = 1
                                if game_over(result_copy) == True:
                                    validation2 = True
                                    break

                        if validation2 == True:
                            break
                    
                    # If none of the tokens are in blocking positions take a random one.
                    if row == 2 and col == 2:
                        result_copy = copy.deepcopy(result)
                        result_copy[row][col] = 1
                        if game_over(result_copy) == False:
                            while True:
                                row = random.randint(0, 2)
                                col = random.randint(0, 2)
                                if result[row][col] == 2:
                                    break
                                
        # If the last token is not 2 check for tokens that are in a blocking position
        # because it's better to pick up a token that is in a clocking position that one in a winning position.
        # If there are none take a random token.
        else:
            for row in range(3):
                for col in range(3):
                    if result[row][col] == 2:
                        result_copy = copy.deepcopy(result)
                        result_copy[row][col] = 1
                        if game_over(result_copy) == True:
                            validation3 = True
                            break

                if validation3 == True:
                    break
            
            # If none of the tokens are in blocking positions take a random one.
            if row == 2 and col == 2:
                while True:
                    row = random.randint(0, 2)
                    col = random.randint(0, 2)
                    if result[row][col] == 2:
                        break
                    
        
    # check for a winning move for the player with tokens value 1
    winning_move = check_winning_move(1, result)
    if winning_move is not None:
        new_row, new_col = winning_move

    else:
        # check for a blocking move for the player with tokens value 1
        blocking_move = check_blocking_move(1, result)

        if blocking_move is not None:
            new_row, new_col = blocking_move

        else:
            # choose a random empty cell
            while True:
                new_row = random.randint(0, 2)
                new_col = random.randint(0, 2)
                if result[new_row][new_col] == 0:
                    break

    # move the token to the new cell
    result[row][col] = 0
    result[new_row][new_col] = 2
    

    # get the cells with the number
    position_1 = get_square_number(row, col)
    position_2 = get_square_new_number(new_row, new_col)
    print(position_1, position_2)


print(result)
```
